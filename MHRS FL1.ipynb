```python
#Rotating Tiff Images
try:
    # Read image from disk.
    img1 = cv2.imread("E:\python\ALPSRP180540490-H2.2_UA\HH-ALPSRP180540490-H2.2_UA.tif")
    (rows, cols) = img1.shape[:2]
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
    res11 = cv2.warpAffine(img1, M, (cols, rows))
    cv2.imwrite('result.tiff', res11)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res11)
```
